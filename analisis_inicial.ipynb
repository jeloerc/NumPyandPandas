Scatter Plots - Relationships between Variables:
-Scatter plots are excellent for visualizing the relationship between two numerical variables.
In a new cell, let's see how sepal length relates to sepal width:

```python
plt.figure(figsize=(8, 6)) # Optional: Adjust figure size
sns.scatterplot(data=df_iris, x='sepal_length', y='sepal_width')
plt.title('Sepal Length vs. Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show() # Strictly optional in notebooks, but good practice
```

Now, let's make it more informative by coloring the points based on the species using the hue parameter:

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_iris, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs. Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

Histograms - Distribution of a Single Variable:
-Histograms show how frequently values occur within specific ranges (bins) for a single numerical variable.
Let's look at the distribution of petal lengths:

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df_iris, x='petal_length', bins=10) # 'bins' controls grouping
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

Let's add a Kernel Density Estimate (KDE) curve, which is a smoothed version of the histogram, and view distributions per species:

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df_iris, x='petal_length', kde=True, hue='species')
plt.title('Distribution of Petal Length by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

Bar Plots - Categorical Data:
-Bar plots are great for comparing counts or aggregated values across different categories.
Let's visualize the count of each species we saw this with value_counts()

```python
plt.figure(figsize=(7, 5))
sns.countplot(data=df_iris, x='species')
plt.title('Number of Samples per Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

Box Plots - Distribution Summary by Category:
-Box plots (or box-and-whisker plots) show the distribution of numerical data across different categories, highlighting the median, quartiles, and potential outliers.
Let's compare the distribution of petal_width for each species:

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_iris, x='species', y='petal_width')
plt.title('Distribution of Petal Width by Species')
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')
plt.show()
```

Pair Plot: Creates a matrix of scatter plots for all combinations of numerical variables, and histograms on the diagonal. Excellent for a quick overview.

```python
# Create a pair plot of all numerical features, colored by species
# This might take a few seconds to render
sns.pairplot(df_iris, hue='species', palette='rocket')
plt.suptitle('Pair Plot of Iris Dataset Features', y=1.02) # Add a main title slightly above
plt.show()
```
